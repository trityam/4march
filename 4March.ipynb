```python
import plotly.graph_objects as go
import seaborn as sns
```

```python
t=sns.load_dataset('titanic')
```

```python
#1)Ans:____
fig=go.Figure(data=[go.Scatter(x=t.age,y=t.fare,mode='markers')])
fig.show()
```

```python
#2)Ans:___
import plotly.express as px

d=px.data.tips()
fig=px.box(d,x='day',y='tip')
fig.show()
```

```python
#3)Ans:___
fig=px.histogram(d,x='sex',y='total_bill',pattern_shape='smoker',color='day')
fig.show()
```

```python
#4)ans__
d=px.data.iris()
fig=px.scatter_matrix(d,dimensions=("sepal_length", "sepal_width", "petal_length", "petal_width"),color='species')
fig.show()
```

```python
#5)Ans:____
import numpy as np
import plotly.figure_factory as ff
d=[np.random.randn(100)]
labels=['distplot']
fig=ff.create_distplot(d,labels)
fig.show()
```
